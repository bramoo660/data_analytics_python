## Data Preprocessing
- Replacing missing values: replace '??' in KM with NaN
- change the column type to the required datatype(from String to float)

```python
# df('KM').replace('??', np.nan, inplace = True)
```

```python
df['KM'] = df['KM'].replace('??', np.nan)
```

```python
df['KM'] = df['KM'].astype(float)
```

```python
print(np.unique(df['KM']))
```

```python
df['HP'] = df['HP'].replace('????', np.nan)
df['HP'] = df['HP'].astype(float)
print(np.unique(df['HP']))
```
