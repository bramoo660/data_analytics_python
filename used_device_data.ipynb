## Filling in Missing Values using mean attrib in following 
- main_camera_mp
- selfie_camera_mp
- int_memory
- ram
- battery
- weight
#### Checking to see if missing values still exit

```python
mean_mcm = device_data['main_camera_mp'].mean()
device_data['main_camera_mp'] = device_data['main_camera_mp'].fillna(mean_mcm)
```

```python
mean_scm = device_data['selfie_camera_mp'].mean()
device_data['selfie_camera_mp'] = device_data['selfie_camera_mp'].fillna(mean_scm)
```

```python
mean_im = device_data['int_memory'].mean()
device_data['int_memory'] = device_data['int_memory'].fillna(mean_im)
```

```python
mean_rm = device_data['ram'].mean()
device_data['ram'] = device_data['ram'].fillna(mean_rm)
```

```python
mean_bt = device_data['battery'].mean()
device_data['battery'] = device_data['battery'].fillna(mean_bt)
```

```python
mean_wt = device_data['weight'].mean()
device_data['weight'] = device_data['weight'].fillna(mean_wt)
```

#### Final Check on Data for Missing Values 

```python
## Display the new data table
device_data.isnull().sum()
```

#### Importing another Pandas Library at top of Import section to Aid in our Descriptive Analysis

```python
plt.scatter(device_data['screen_size'], device_data['used_price'], c = 'blue')

plt.title('Relationship Between Screen size and Price')

plt.ylabel('Price (Dollars)')

plt.xlabel('screen size (Diagonal)')

plt.savefig('Screen_price_rel.png')
plt.show
```
