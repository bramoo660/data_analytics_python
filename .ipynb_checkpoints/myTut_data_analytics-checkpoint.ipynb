```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy as sp
```

```python
## intersection, differentiation, neither

arr1 = np.array([1,2,3,4,5])
arr2 = np.array([3,4,5,6,7])
print(np.intersect1d(arr1, arr2))
print(np.union1d(arr1, arr2))
print(np.setdiff1d(arr1, arr2))
print(np.setxor1d(arr1, arr2))
```

```python
## os, int_memory, days_used
data_df = data_df.drop(['os', 'int_memory', 'days_used'], axis = 1)
data_df.head(5)
```

```python
# renaming column headings

data_df = data_df.rename(columns={"brand_name": "brand", "screen_size": "size", "release_year": "year"})
data_df.head(5)
```

```python
## Total number Rows and Columns

data_df.shape
```

```python
# checking for rows with duplicate data

duplicate_data_df = data_df[data_df.duplicated()]
print("Number of rows with duplicate: ", duplicate_data_df.shape)
```

```python
## count the rows before removing data

data_df.count()
```

```python
## drop the duplicate

data_df = data_df.drop_duplicates()
data_df.head()
```
